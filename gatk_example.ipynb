```python
%%bash
# подготовим vcf для анализа
for SAM_FILE in brca_*;
do
#echo $SAM_FILE ;
#echo ${SAM_FILE:0:-4};
FILE_NAME=${SAM_FILE:0:-4} # создаем имя файла
# echo $FILE_NAME
samtools view -b -h -o output/${FILE_NAME}.bam $SAM_FILE # создаем bam

# сортируем и индексируем bam
samtools sort -o output/$FILE_NAME_sorted.bam output/$FILE_NAME.bam
samtools index output/$FILE_NAME_sorted.bam

# gvcf
gatk HaplotypeCaller -I output/$FILE_NAME_sorted.bam -R reference/chr17.fasta -ERC GVCF --output output/$FILE_NAME.g.vcf

done

# проверим есть ли файлы
ls output | grep g.vcf
```

```python
!gatk CombineGVCFs -R reference/chr17.fasta -V output/brca_1.g.vcf -V output/brca_2.g.vcf -O output/cohort.g.vcf.gz
```

### Пример результата исследований

Выполнен анализ наличия мутаций в кодирующих и прилегающих областях генов BRCA1/BRCA2 (RefSeq: NCBI Reference Sequence Database - NM_007294.3, NM_000059.4).

Исследование проводилось с использованием набора реагентов "СОЛО-ТЕСТ АВС 48С-А" (РУ № РЗН 2020/9943) на приборе Illumina MiSeq методом высокопроизводительного секвенирования по ТУ 21.20.23-001-91709359-2018.

Диагностические характеристики запуска:

Количество прочтений на образец - 323037
Прочитано оснований - 39437925
Среднее значение покрытия образца - 440,3
